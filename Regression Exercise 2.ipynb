```python
%matplotlib inline
import numpy as np
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
# http://scikit-Learn.org/stable/modules/generated/sklearn.datasets.Load_tris.html
iris = datasets.load_iris()
list(iris.keys())
```

Probability that the measured sepal length of 4.8 cm is an iris virginica:

```python
X = iris["data"][:, 0:1] # sepal length array
y = (iris["target"] == 2).astype(np.int64) # array of 1 if Iris-Virginica, else 0

log_reg = LogisticRegression()
log_reg.fit(X, y)

X_new = np.linspace(4, 8, 1000).reshape(-1, 1)
y_proba = log_reg.predict_proba(X_new)
plt.plot (X_new, y_proba[:, 1], "g-", label="Iris-Virginica")
plt.plot (X_new, y_proba[:, 0], "b--", label="Not Iris-Virginica")

proba_sepal_length = log_reg.predict_proba([[4.8]])
proba_sepal_length = proba_sepal_length[0,1]
proba_sepal_length
```

Probability that the measured sepal width of 2.5 cm is an iris virginica:

```python
X = iris["data"][:, 1:2] # sepal width array
y = (iris["target"] == 2).astype(np.int64) # array of 1 if Iris-Virginica, else 0

log_reg = LogisticRegression()
log_reg.fit(X, y)

X_new = np.linspace(0, 5, 1000).reshape(-1, 1)
y_proba = log_reg.predict_proba(X_new)
plt.plot (X_new, y_proba[:, 1], "g-", label="Iris-Virginica")
plt.plot (X_new, y_proba[:, 0], "b--", label="Not Iris-Virginica")

proba_sepal_width = log_reg.predict_proba([[2.5]])
proba_sepal_width = proba_sepal_width[0,1]
proba_sepal_width
```

Probability that the measured petal length of 5.3 cm is an iris virginica:

```python
X = iris["data"][:, 2:3] # petal length array
y = (iris["target"] == 2).astype(np.int64) # array of 1 if Iris-Virginica, else 0

log_reg = LogisticRegression()
log_reg.fit(X, y)

X_new = np.linspace(2, 7, 1000).reshape(-1, 1)
y_proba = log_reg.predict_proba(X_new)
plt.plot (X_new, y_proba[:, 1], "g-", label="Iris-Virginica")
plt.plot (X_new, y_proba[:, 0], "b--", label="Not Iris-Virginica")

proba_petal_length = log_reg.predict_proba([[5.3]])
proba_petal_length = proba_petal_length[0,1]
proba_petal_length
```

Probability that the measured petal width of 2.4 cm is an iris virginica:

```python
X = iris["data"][:, 3:] # petal width array
y = (iris["target"] == 2).astype(np.int64) # array of 1 if Iris-Virginica, else 0

log_reg = LogisticRegression()
log_reg.fit(X, y)

X_new = np.linspace(0, 3, 1000).reshape(-1, 1)
y_proba = log_reg.predict_proba(X_new)
plt.plot (X_new, y_proba[:, 1], "g-", label="Iris-Virginica")
plt.plot (X_new, y_proba[:, 0], "b--", label="Not Iris-Virginica")

proba_petal_width = log_reg.predict_proba([[2.4]])
proba_petal_width = proba_petal_width[0,1]
proba_petal_width
```

Combined probability of all measured attributes that the iris is an iris virginica:

```python
import math

z = 0.25*proba_sepal_length+0.25*proba_sepal_width+0.25*proba_petal_length+0.25*proba_petal_width
# Assumption: All attributes are weighted equally to one quarter

proba = 1/(1+math.exp(-z))
proba = round(proba*100, 1)

print("The probability that an iris with the parameters [4.8, 2.5, 5.3, 2.4] is an Iris-Virginica =", proba, "%")
```
